###  Apply: Transformaciones en base a una columna

```python
df_aviones_raw = pd.read_csv("./data/dataset_dirty_aviones.csv", index_col = "Id_vuelo")
```

Nos piden que creemos una columna nueva para clasificar los vuelos en tres categorias:
* LD -> Larga Distancia distancia > 10000Km
* MD -> Media Distacia distancia entres 2000Km y 10000Km
* CD -> Distancia corta -> distancias < 2000 Km

```python
# Solucion "antipattern"
clasificacion = []
for distancia in df_aviones["Distancia"]:
    if distancia > 10000:
        clasificacion.append("LD")
    elif distancia >= 2000:
        clasificacion.append("MD")
    else:
        clasificacion.append("CD")
df_aviones_raw["categoria_vuelo"] = clasificacion
```

```python
# Con una función 
def clasificador_distancia(serie):
    clasificacion = []
    for distancia in serie:
        if distancia > 10000:
            clasificacion.append("LD")
        elif distancia >= 2000:
            clasificacion.append("MD")
        elif distancia > 500:
            clasificacion.append("CD+")
        else:
            clasificacion.append("CD")
    return clasificacion

df_aviones_raw["categoria_vuelo"] = clasificador_distancia(df_aviones_raw["Distancia"])
```

```python
df_aviones_raw.value_counts("categoria_vuelo")
```

#### Apply

Es un método de las `Series` y de los `DataFrame` que sirve para aplicar funciones valor a valor.  
Apply devuelve el resultado que devuelva la función para cada valor agrupándolo en una Serie con los mismos indices que la serie de entrada (la que invoca el apply)


```python
def clasificador_apply(distancia):
    if distancia > 10000:
            clasificacion = "LD"
    elif distancia >= 2000:
        clasificacion = "MD"
    else:
        clasificacion = "CD"
    return clasificacion
```

```python
df_aviones_raw["Distancia"].apply(clasificador_apply)
```

Considerando el consumo (consumo_kg):
* Para mayorers de 8000, categoria C
* Para consumos entre 5000 y 8000, categoria B
* Para consumos menores de 5000, categoria A

```python
def cat_consumo(consumo):
    if consumo > 8000:
            clasificacion = "C"
    elif consumo >= 5000:
        clasificacion = "B"
    else:
        clasificacion = "A"
    return clasificacion
```

`df_aviones_raw["consumo_kg"].apply(cat_consumo)` No se puede realizar porque los datos están sucios, hay "," en vez de "." para los decimales


```python
def reemplaza(consumo): # Se crea otra función para reemplazar los valores
    if type(consumo)== str:
        return float(consumo.replace("," , "."))
    else:
        return consumo
df_aviones_raw["consumo_kg"] = df_aviones_raw["consumo_kg"].apply(reemplaza)
```

```python
df_aviones_raw["consumo_kg"] = df_aviones_raw["consumo_kg"].apply(cat_consumo) # Ahora si!
```

```python
df_aviones_raw
```

##  Apply: Transformaciones sobre varias columnas y sobre selecciones/filtrados

```python
df_aviones = pd.read_csv("./data/dataset_inicial_aviones.csv", index_col = "Id_vuelo")
df_aviones_2 = pd.read_csv("./data/dataset_apply_aviones.csv", index_col = "Id_vuelo")
```

Nos piden clasificar los vuelos según las siguientes reglas:  

(Consumo por kilometro es consumo_kg/distancia)   

* Para vuelos de > 10000Km:
    * Si su el consumo por kilometro es mayor que 11 y la duracion menos de 1000 minutos, cat: MC (muy contaminante)
    * Si su el consumo por kilometro es mayor que 11 y la duracion más de 1000 minutos o su consumo es menor que 11, cat: AC (altamente contaminante)
* Para vuelos de < 10000km:
    * Si su consumo por kilometro es mayor que 10, cat: MC
    * Si su consumo por kilometro es menor que 10, y su duracion menor que 600: AC
    * En cualquier otro caso: C (Contaminante) 

Para utilizar varias colimnas hay que cambiar la función para que todos los datos lleguen en un solo argumento

```python
def cat_contaminacion(row):
    distancia =row["Distancia"]
    consumo_kg = row["consumo_kg"]
    duracion = row["duracion"]
    consumo_km = consumo_kg/distancia
    if distancia > 10000:
        if consumo_km > 11 and duracion > 1000:
            categoria = "MC"
        else:
            categoria = "AC"
    elif distancia < 10000:
        if consumo_km > 10:
            categoria = "MC"
        elif duracion < 600:
            categoria = "AC"
        else:
            categoria = "C"
    return categoria
```

Al hacer `.apply`, muy importante añadir `axis = 1` para operar en varias columnas.

```python
df_aviones["Cat_contaminacion"] = df_aviones[["duracion", "consumo_kg", "Distancia"]].apply(cat_contaminacion, axis = 1) 
```

```python
df_aviones
```

Para terminar, corrijamos el `DataFrame` df_aviones_2:  
Dos compañías, Airnar y MoldaviAir, han entregado mal sus datos. Después de preguntarles nos confirman que ambas por error han utilizado una codificación antigüa. Hay que quitar la "d:" y dividir entre 60 para tener la duración correcta. Pues nada construyamos la función y luego se la aplicamos (apply) a la columna `duracion`...

```python
def corrige_duracion(row):
    new_row = int(row.replace("d:", "")) / 60
    return int(new_row)
```

```python
es_bad_company = df_aviones_2["Aircompany"].isin(["Airnar", "MoldaviAir"]) # La condición
```

```python
df_aviones_2.loc[es_bad_company, "duracion"] = df_aviones_2.loc[es_bad_company, "duracion"].apply(corrige_duracion)
```

##  Groupby

```python
df_aviones = pd.read_csv("./data/dataset_inicial_aviones.csv", index_col = "Id_vuelo")
```

```python
for compania in df_aviones["Aircompany"].unique():
    es_compania = df_aviones["Aircompany"] == compania
    print(f"Para la compañía {compania}:")
    for avion in df_aviones["avion"].unique():
        es_avion = df_aviones["avion"] == avion
        consumo = df_aviones.loc[es_compania & es_avion,"consumo_kg"].mean()
        print(f"Tipo <{avion}> consumo medio por vuelo <{consumo:.2f}>")
```

En vez de hacer este bucle para calcular el consumo medio por avión de cada compañía, se puede utilizar el método `groupby(["primera_columna","segunda_columna"])[columna_a_aplicar].metodo_aplicar`

```python
df_aviones.groupby(["Aircompany", "avion"])["consumo_kg"].mean()
```

Para operar sobre varias columnas hay que hacer una lista dentro de una lista `[[]]`

```python
df_aviones.groupby(["Aircompany", "Origen","Destino"])[["consumo_kg", "duracion"]].mean()
```

Son muchos resultados y el DF no aparece completo, para filtrarlo, guardarlo en una variable y acceder con el metodo `.loc[]`

```python
df_medias = df_aviones.groupby(["Aircompany", "Origen","Destino"])[["consumo_kg", "duracion"]].mean()

df_medias.loc["Airnar"]
```

```python
# Resultados para todos los vuelos París-Cádiz
df_medias.loc[:,"París", "Cádiz"]

# Resultados para los vuelos a Ginebra y Nueva York de FlyQ
df_medias.loc["FlyQ", : , ["Ginebra", "Nueva York"]]
```

Nosotros, en general, si queremos conservar la salida de un groupby para seguir procesándola, emplearemos el argumento as_index con valor "False"



```python
resultado_no_index = df_aviones.groupby(["Aircompany","Origen","Destino"], as_index = False)[["consumo_kg","duracion"]].mean()
```

### Agg: Funciones de usuario


Supongamos que ahora queremos obtener por Compañía y tipo de Avión, el número de vuelos y además el consumo medio y el destino más visitado. Hala ahí es nada. Descompongámoslo en tres pequeños problemas y luego veremos como juntarlos con la ayuda del método `agg`

1. Por compañia y tipo de avión, el número de vuelos. Ese es fácil, lo hicimos en la píldora anterior:

```python
df_grouped = df_aviones.groupby(["Aircompany", "avion"]) #----> Para evitar repetir esto todo el rato, meterlo en una variable
df_grouped["Origen"].count()
```

2. Por compañia y tipo de avión, el consumo medio. También lo tengo:

```python
df_grouped["consumo_kg"].mean()
```

3. Por compañía y tipo de avion el destino más visitado.

```python
def mas_visitado(serie):
    return serie.mode()[0] #Si hay varios destinos empatados, coger el primero
```

```python
df_grouped["Destino"].apply(mas_visitado)
```

### Agg, aggregate: aplicando funciones y operaciones diferentes por columnas

Muy sencillo, vamos a usar el método agg o el aggregate al que se le pasa un diccionario:  
{nombre_de_la_columna: operacion_sobre_esa_columna,  
 nombre_de_la_columna2: operacion_sobre_esa_columna,...  
 nombre_de_la_columnan: operacion_sobre_esa_columna}

```python
agregaciones = {
    "Origen" : len,  #--> Aquí no vale.count(), ha de ser len.
    "consumo_kg" : "mean", #---> Para la media en agregaciones hay que usar una etiqueta (string)
    "Destino" : mas_visitado #---> Aquí si la función creada antes
}

df_grouped[["Origen", "consumo_kg", "Destino"]].agg(agregaciones)
```

### Transform fuera del *groupby*


Aunque lo vamos a ver en el contexto de las agrupaciones hechas con "groupby", `transform` es un método que también se aplica a `Series` y `DataFrame` sin que haya un `groupby` de por medio. 

Es similar a `apply`, y sus principales diferencias son:
* Puede aplicar una o varias funciones (sí varias funciones a la vez, introducidas como lista o diccionario)
* Sólo se puede aplicar a una serie (o columna) a la vez (no vale el método de varias columnas)

```python
def a_mayusculas(valor):
    return valor.upper()

def abrevia(valor):
    return valor[0:3]
```

```python
df_aviones["Destino"].transform([a_mayusculas, abrevia])
```

Se puede hacer a partir de un diccionario:

```python

transformaciones = {
    "Des_Mayuscula" : a_mayusculas,
    "Des_Abrev" : abrevia
}
df_aviones[["Destino", "Des_Abrev"]] = df_aviones["Destino"].transform(transformaciones)
df_aviones
```

### Transform para agrupaciones

```python
### Sin transform
df_aviones.groupby(["Aircompany", "avion"])["duracion"].mean()
```

```python
### Con transform
df_aviones.groupby(["Aircompany", "avion"])["duracion"].transform("mean")
```

```python
df_aviones["media_duracion"] = df_aviones.groupby(["Aircompany", "avion"])["duracion"].transform("mean")
df_aviones
```

```python
df_aviones["desviacion_duracion"] = df_aviones["duracion"] / df_aviones["media_duracion"]
df_aviones
```
