### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center").
2. Histograma del área total ocupada por las ciudades.
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
df_california[["city", "population_total"]].sort_values(by="population_total" ,ascending=False)
```

```python
#1
ciudades = ["LosAngeles", "SanDiego", "Sacramento", "Anaheim", "SantaAna"]
valores = [3884307, 1345895, 466488, 336000, 324528]
colores = ["mediumslateblue", "darkgrey", "lightcoral", "darkkhaki", "rosybrown"]
```

```python
plt.figure(figsize=(5, 5))

plt.bar(ciudades, valores, color = colores)
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.title("Ciudades más pobladas de california");
```

```python
fonts = {
    "size" : 14,
    "family" : "Bradley Hand itc"
}
```

```python
#2
plt.figure(figsize=(5, 5))
plt.hist(df_california['area_total_km2'], bins= 100, color='steelblue', alpha=0.7, label = "hist area")
plt.xlabel('Superficie ($km^2$)', fontdict= fonts)
plt.ylabel('Número de ciudades', fontdict= fonts)
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.legend();
```

```python
max_x_value = df_california.area_water_km2.idxmax()
max_y_value = df_california.population_total.max()
```

```python

#3
plt.scatter(df_california["area_water_km2"], df_california["population_total"], c = "g", s= 10 * df_california["city"].value_counts(), alpha = 0.5)
plt.xlabel("Área de agua en la ciudad ($km^2$)")
plt.ylabel("Población (miles de habitantes)")
plt.title("Población por kilómetro cuadrado de agua")

props_flecha = {
    "arrowstyle" : "<->",
    "connectionstyle" : "angle3"
}
plt.annotate("Outliniers", xy = (max_x_value, max_y_value), xytext = (max_x_value, 0.5), arrowprops = props_flecha);

#mmmmm no se por que no me sale el anotate....
```

```python
#4
plt.boxplot(df_california.elevation_m, vert= False, patch_artist = True,
           boxprops = dict(facecolor = "red"),  medianprops = dict(color = "black"))
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la altitud de las ciudades de California");
```

```python
# Unir los 4:

# Gráfico 1
plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.bar(ciudades, valores, color=colores)
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.title("Ciudades más pobladas de California")

# Gráfico 2
plt.subplot(2, 2, 2)
fonts = {"size": 14, "family": "Bradley Hand ITC"}
plt.hist(df_california['area_total_km2'], bins=100, color='steelblue', alpha=0.7, label="Hist Area")
plt.xlabel('Superficie ($km^2$)', fontdict=fonts)
plt.ylabel('Número de ciudades', fontdict=fonts)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend()

# Gráfico 3
plt.subplot(2, 2, 3)
plt.scatter(df_california["area_water_km2"], df_california["population_total"], c="g", s=10 * df_california["city"].value_counts(), alpha=0.5)
plt.xlabel("Área de agua en la ciudad ($km^2$)")
plt.ylabel("Población (miles de habitantes)")
plt.title("Población por kilómetro cuadrado de agua")

# Gráfico 4
plt.subplot(2, 2, 4)
plt.boxplot(df_california.elevation_m, vert=False, patch_artist=True, boxprops=dict(facecolor="red"), medianprops=dict(color="black"))
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la altitud de las ciudades de California")

plt.tight_layout()
plt.savefig('graficos_combinados_California.png')
plt.show();
```
